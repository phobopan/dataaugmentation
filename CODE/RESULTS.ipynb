```python
# Create a figure and axis object
fig, ax1 = plt.subplots()

# Plot RMSE and MAE on the left y-axis
ax1.set_xlabel('Resampling Ratio')
ax1.set_ylabel('µg/m3')
ax1.plot(r, rmse_12, 'o-', label='RMSE', color='blue')
ax1.plot(r, mae_12, 's-', label='MAE', color='green')
ax1.legend(loc='upper left')
ax1.tick_params(axis='y')

# Create a second y-axis for R²
ax2 = ax1.twinx()
ax2.set_ylabel('R²')
ax2.plot(r, r2_12, 'd-', label='R²', color='red')
ax2.legend(loc='upper right')
ax2.tick_params(axis='y')

# Add title and grid
plt.title('RMSE, MAE, and R² across multiple resampling ratios - 12.1 Threshold - Whole')
#ax1.grid(True)

# Show plot
plt.show()
```
